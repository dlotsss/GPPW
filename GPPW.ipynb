```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data: 1 = accessible (dark green), 0 = not accessible (light green)
data = [
    [1, 0, 1, 1, 1, 1, 1],  # Entrance & Ramp
    [1, 0, 0, 0, 1, 1, 1],  # Elevator
    [0, 0, 0, 0, 0, 1, 1],  # Restroom
]

buildings = [
    'City Mall', 'Abzal', 'Arbat', 'Central Department Store (ЦУМ)', 'NIS ChB', 'Clinic #11', 'Clinic of city hospital №1'
]
features = [
    'Entrance & Ramp', 'Elevator', 'Restroom'
]

plt.figure(figsize=(15, 5))
sns.heatmap(
    data,
    annot=True,
    fmt='d',
    cmap=sns.color_palette(["#b7efc5", "#006400"]),  # light green, dark green
    cbar=False,
    xticklabels=buildings,
    yticklabels=features
)
plt.title('Accessibility Heatmap of Buildings in Karaganda')
plt.tight_layout()
plt.show()
```

The heatmap displays three key accessibility features (Entrance & Ramp, Elevator, and Restroom) across seven Karaganda locations (shopping centers, school and clinics), using 1 to indicate presence and 0 absence. It shows that ramp access is available at all sites except Abzal Shopping Centre, while elevators are provided at only four locations (City Mall, NIS ChB Karaganda, Clinic №11, and City Hospital’s Clinic №1) and missing at Abzal, Arbat, and the Central Department Store. Accessible restrooms are even rarer, appearing only Clinic №11, and City Hospital’s Clinic №1, and absent at the other five buildings.
